# Step 1: Import Libraries


```python
# Importing essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

# Step 2: Load the Dataset

```python
# Load dataset
df = pd.read_csv('customer_purchase_data.csv')

# Display the first few rows of the dataset
df.head()
```

# Summary Statistics

```python
# Summary statistics for numerical columns
df.describe()
```

# Distribution of Target Variable (PurchaseStatus)

```python
# Plot the distribution of the target variable (PurchaseStatus)
sns.countplot(x='PurchaseStatus', data=df)
plt.title('Distribution of Purchase Status')
plt.show()
```

# Correlation Matrix

```python
# Calculate correlation matrix
corr_matrix = df.corr()

# Plot correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

# Age and Annual Income vs. Purchase Status

```python
# Age vs Purchase Status
plt.figure(figsize=(10, 6))
sns.boxplot(x='PurchaseStatus', y='Age', data=df)
plt.title('Age vs Purchase Status')
plt.show()

# Annual Income vs Purchase Status
plt.figure(figsize=(10, 6))
sns.boxplot(x='PurchaseStatus', y='AnnualIncome', data=df)
plt.title('Annual Income vs Purchase Status')
plt.show()
```

# Time Spent on Website and Purchases

```python
# Scatter plot between TimeSpentOnWebsite and NumberOfPurchases
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TimeSpentOnWebsite', y='NumberOfPurchases', hue='PurchaseStatus', data=df)
plt.title('Time Spent on Website vs Number of Purchases')
plt.show()
```

# Pairplot for All Features

```python
# Pairplot for numerical features
sns.pairplot(df[['Age', 'AnnualIncome', 'NumberOfPurchases', 'TimeSpentOnWebsite', 'PurchaseStatus']], hue='PurchaseStatus')
plt.show()
```

# Data Preprocessing

```python
print(df.isnull().sum())
```

# Split the data into features and target variable

```python
X = df.drop('PurchaseStatus', axis=1)
y = df['PurchaseStatus']
```

# Scaling the features

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Split the data into training and testing sets

```python
x_train, x_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# Models

# Decission Tree Classifier

```python
# Decision Tree Classifier
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(x_train, y_train)
y_pred_dt = dt_model.predict(x_test)
```

```python
# Decision Tree Evaluation
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
recall_dt = recall_score(y_test, y_pred_dt)
f1_dt = f1_score(y_test, y_pred_dt)
cm_dt = confusion_matrix(y_test, y_pred_dt)
```

```python
print(f'Decision Tree Classifier:\n Accuracy: {accuracy_dt}\n Precision: {precision_dt}\n Recall: {recall_dt}\n F1 Score: {f1_dt}')
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Blues')
plt.title('Decision Tree Confusion Matrix')
plt.show()
```

# Logistic Regression

```python
# Logistic Regression
lr_model = LogisticRegression(random_state=42)
lr_model.fit(x_train, y_train)
y_pred_lr = lr_model.predict(x_test)
```

```python
# Logistic regression Evaluation
accuracy_lr = accuracy_score(y_test, y_pred_lr)
precision_lr = precision_score(y_test, y_pred_lr)
recall_lr = recall_score(y_test, y_pred_lr)
f1_lr = f1_score(y_test, y_pred_lr)
cm_lr = confusion_matrix(y_test, y_pred_lr)
```

```python
print(f'Logistic Regression:\n Accuracy: {accuracy_lr}\n Precision: {precision_lr}\n Recall: {recall_lr}\n F1 Score: {f1_lr}')
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues')
plt.title('Logistic Regression Confusion Matrix')
plt.show()
```

# Suppor Vector Machine

```python
# Support Vector Machine
svm_model = SVC(random_state=42)
svm_model.fit(x_train, y_train)
y_pred_svm = svm_model.predict(x_test)
```

```python
# SVM Evaluation
accuracy_svm = accuracy_score(y_test, y_pred_svm)
precision_svm = precision_score(y_test, y_pred_svm)
recall_svm = recall_score(y_test, y_pred_svm)
f1_svm = f1_score(y_test, y_pred_svm)
cm_svm = confusion_matrix(y_test, y_pred_svm)
```

```python
print(f'Support Vector Machine:\n Accuracy: {accuracy_svm}\n Precision: {precision_svm}\n Recall: {recall_svm}\n F1 Score: {f1_svm}')
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Blues')
plt.title('Support Vector Machine Confusion Matrix')
plt.show()
```

# Random Forest

```python
# Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train, y_train)
y_pred_rf = rf_model.predict(x_test)
```

```python
# Random Forest Evaluation
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)
cm_rf = confusion_matrix(y_test, y_pred_rf)
```

```python
print(f'Random Forest:\n Accuracy: {accuracy_rf}\n Precision: {precision_rf}\n Recall: {recall_rf}\n F1 Score: {f1_rf}')
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues')
plt.title('Random Forest Confusion Matrix')
plt.show()
```

# Evaluation for each model

```python
# Evaluation Summary for Decision Tree
print(f'Decision Tree Classifier Evaluation:')
print(f'Accuracy: {accuracy_dt:.4f}')
print(f'Precision: {precision_dt:.4f}')
print(f'Recall: {recall_dt:.4f}')
print(f'F1 Score: {f1_dt:.4f}')
print('-' * 50)

# Evaluation Summary for Logistic Regression
print(f'Logistic Regression Evaluation:')
print(f'Accuracy: {accuracy_lr:.4f}')
print(f'Precision: {precision_lr:.4f}')
print(f'Recall: {recall_lr:.4f}')
print(f'F1 Score: {f1_lr:.4f}')
print('-' * 50)

# Evaluation Summary for Support Vector Machine
print(f'Support Vector Machine Evaluation:')
print(f'Accuracy: {accuracy_svm:.4f}')
print(f'Precision: {precision_svm:.4f}')
print(f'Recall: {recall_svm:.4f}')
print(f'F1 Score: {f1_svm:.4f}')
print('-' * 50)

# Evaluation Summary for Random Forest
print(f'Random Forest Classifier Evaluation:')
print(f'Accuracy: {accuracy_rf:.4f}')
print(f'Precision: {precision_rf:.4f}')
print(f'Recall: {recall_rf:.4f}')
print(f'F1 Score: {f1_rf:.4f}')
print('-' * 50)

# Classification Reports
print("Decision Tree Classification Report:")
print(classification_report(y_test, y_pred_dt))

print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_lr))

print("Support Vector Machine Classification Report:")
print(classification_report(y_test, y_pred_svm))

print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
```

# Compare Models

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Store the evaluation metrics in a dictionary for comparison
metrics = {
    'Model': ['Decision Tree', 'Logistic Regression', 'Support Vector Machine', 'Random Forest'],
    'Accuracy': [accuracy_dt, accuracy_lr, accuracy_svm, accuracy_rf],
    'Precision': [precision_dt, precision_lr, precision_svm, precision_rf],
    'Recall': [recall_dt, recall_lr, recall_svm, recall_rf],
    'F1 Score': [f1_dt, f1_lr, f1_svm, f1_rf]
}

# Convert the metrics into a DataFrame for easy handling
results = pd.DataFrame(metrics)

# Define color scheme for models
colors = ['#FF6347', '#FFD700', '#3CB371', '#6495ED']

# Plotting the comparison for all metrics
fig, axes = plt.subplots(1, 4, figsize=(24, 6))

# Bar plots for each metric
for i, metric in enumerate(['Accuracy', 'Precision', 'Recall', 'F1 Score']):
    axes[i].bar(results['Model'], results[metric], color=colors)
    axes[i].set_title(f'Comparison of {metric}')
    axes[i].set_ylabel(metric)
    axes[i].set_ylim(0, 1)  # Since metrics are between 0 and 1
    axes[i].set_xlabel('Model')
    axes[i].set_xticklabels(results['Model'], rotation=15)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
